**$ 2^{nd} $ order ODE Solution by RK4** \\


---


The ODE is $ \frac{d^2y}{dx^2}=x\left(\frac{dy}{dx}\right)^2-y^2 $


spliting it into two coupled first order equations: \\

$ \frac{dy}{dx} = z = f(x,y,z) $ (say) \\
$ \frac{dz}{dx} = xz^2 - y^2 $ \\

Initial Conditions:
$ y=1 , y' = z = 0 $ when $ x=0 $

```python
import numpy as np

#Initial values
x0,y0,z0 = 0,1,0
#Increment in x 
h = 0.2
def f(x,y,z): return z
def g(x,y,z):
  return x*z**2 - y**2
y = y0
z = z0
x = 0.2
X,Y,Z = [],[],[]

for i in np.arange(x0,x+h,h):
  X.append(i)
  Y.append(y)
  Z.append(z)
  
  K1 = h*f(i,y,z)
  L1 = h*g(i,y,z)
  
  K2 = h*f(i+h/2,y+K1/2,z+L1/2)
  L2 = h*g(i+h/2,y+K1/2,z+L1/2)
  
  K3 = h*f(i+h/2,y+K2/2,z+L2/2)
  L3 = h*g(i+h/2,y+K2/2,z+L2/2)
  
  K4 = h*f(i+h,y+K3,z+L3)
  L4 = h*g(i+h,y+K3,z+L3)
  
  y += (K1+2*K2+2*K3+K4)/6
  z += (L1+2*L2+2*L3+L4)/6
print(X)
print(Y)
print(Z)
import matplotlib.pyplot as plt
plt.plot(X,Y)
```

The Damped Harmonic Oscillator

---


The ODE is $ \frac{d^2y}{dx^2}+Î»\frac{dy}{dx}+ky = 0 $

```python
import numpy as np

#Initial values
x0,y0,z0 = 0,0,1
#Increment in x 
h = 0.1
lam,k = 0.5,2
def f(x,y,z): return z
def g(x,y,z):
  return -lam*z-k*y
y = y0
z = z0
x = 20
X,Y,Z = [],[],[]

for i in np.arange(x0,x+h,h):
  X.append(i)
  Y.append(y)
  Z.append(z)
  
  K1 = h*f(i,y,z)
  L1 = h*g(i,y,z)
  
  K2 = h*f(i+h/2,y+K1/2,z+L1/2)
  L2 = h*g(i+h/2,y+K1/2,z+L1/2)
  
  K3 = h*f(i+h/2,y+K2/2,z+L2/2)
  L3 = h*g(i+h/2,y+K2/2,z+L2/2)
  
  K4 = h*f(i+h,y+K3,z+L3)
  L4 = h*g(i+h,y+K3,z+L3)
  
  y += (K1+2*K2+2*K3+K4)/6
  z += (L1+2*L2+2*L3+L4)/6
print(X)
print(Y)
print(Z)
#for plotting
import matplotlib.pyplot as plt
plt.plot(X,Y,'g',label='y')
plt.plot(X,Z,'orange',label=r"$ \frac{dy}{dx}$")
plt.grid()
plt.legend()
plt.show()
```
