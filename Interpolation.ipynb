###Lagrange Interpolation formula

```python
def LI(x,y,X):
    S = 0
    for i in range(len(x)): #i values are = 0 to n-1. Total n values
        L = 1
        for j in range(len(x)): #j values are = 0 to n-1. Total n values
            if i != j: #when i = j the loop breaks and the next loop starts
                L *= (X-x[j])/(x[i]-x[j])
        S += L*y[i]
    return S
```

Interpolation with forward difference
###Newton’s Interpolation formula

```python
def NFDI(x,y,X):
    h = x[1]-x[0]
    C = (X-x[0])/h
    P = C
    f = y[0]  #because we are using forward differences
    k = 1 #this is for changing of P for different terms
    print('Newton’s Forward Difference Table:')
    for i in range(len(x),1,-1): #step is -1 because the loop will run backward(from n to 1). n is len(x) or len(y)
        #can't run this loop from 0 to n beacuse the other loop will not run for (i-1) when i=0
        #so we start this loop from backside but the same no. of times(n-1). From n to 1(n-1 times)
        z = [] #this is to change the y list by the difference table
        for j in range(i-1): #j value is 0,1,...,i-2
            t = y[j+1] - y[j]
            z.append(t)
        y = z #changing the y list by the jth order difference table
        print(y) #printing the table
        f += P*y[0] #taking the 1st term of the jth order difference table
        P *= (C-k)/(k+1) #updating the P
        k += 1
    return f
```

```python
#Lagrange Interpolation
x = [10,12,14,16,18,20,22]
y = [46,66,81,93,77,53,36]
X = 15 # we want to estimate f(X)
Y = LI(x,y,X)
print('The Interpolated value, f('+str(X)+') = ',Y)
# for visualization 
import matplotlib.pyplot as plt
x.append(X)
y.append(Y)
plt.plot(x,y)
plt.show()
```

```python
#Newton's Interpolation
x = [10,12,14,16,18,20,22]
y = [46,66,81,93,77,53,36]
X = 15 # we want to estimate f(X)
Y = NFDI(x,y,X)
print('The Interpolated value, f('+str(X)+') = ',Y)
# for visualization 
import matplotlib.pyplot as plt
x.append(X)
y.append(Y)
plt.plot(x,y)
plt.show()
```
