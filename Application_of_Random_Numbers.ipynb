# Monte Carlo Integration

In 1D:

$$ \int_a^b f(x)dx = \frac{(b-a)}{N} \Sigma_{i=1}^N f(x_i) + error $$

```python
def MCI(f,a,b,N):
    s = 0
    for i in range(N):
        x = random.uniform(a,b)
        s += f(x)
    mean = s/N
    I = (b -a) * mean
    return I
#We can also use numpy module and this will look easy
#but using numpy for generating random numbers might cause problem
#in some specific question involving log()
#depends on which log we're using math. or numpy.
```

```python
def MCI(f,a,b,N):
    x = np.random.uniform(a,b,N)
    mean = sum(f(x))/N
    I = (b -a) * mean
    return I
#This definition is faster
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

#The error function
def erf(x):
    f = lambda t: np.exp(-t**2)
    a, b, N = 0, x, 100000
    I = (2/np.sqrt(np.pi)) * MCI(f,a,b,N)
    return I

ERF = []
X = []
x = -3
while x <= 3:
    X.append(x)
    ERF.append(erf(x))
    x += 0.1
plt.plot(X,ERF)
plt.grid()
plt.title('The Error Function')
plt.xlabel('$x$')
plt.ylabel('$erf(x)$')
plt.show()
```

### 4. Use the Python function of prob-1 to calculate the following integration
$$ g(x)= 1 - erf\left(\frac{1}{2 \sqrt{x}}\right) $$
### for $ 0.1 \le x \le 3 $ and $ \delta x = 0.1 $. Plot $g(x)$ $ vs $ $x $ within the given range.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

def g(x):
    f = lambda t: np.exp(-t**2)
    a, b, N = 0, 1/(2*x**0.5), 100000
    I = (2/np.sqrt(np.pi)) * MCI(f,a,b,N)
    return 1- I

G = []
X = []
x = 0.1
while x <= 3:
    X.append(x)
    G.append(g(x))
    x += 0.1
plt.plot(X,G,'r-')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

#The complementary error function
def erfc(x):
    f = lambda t: np.exp(-t**2)
    a, b, N = 0, x, 100000
    I = (2/np.sqrt(np.pi)) * MCI(f,a,b,N)
    return 1- I

ERFC = []
X = []
x = -3
while x <= 3:
    X.append(x)
    ERFC.append(erfc(x))
    x += 0.1
plt.plot(X,ERFC,'r-')
plt.grid()
plt.title('The Complementary Error Function')
plt.xlabel('$x$')
plt.ylabel('$erfc(x)$')
plt.show()
```

### 6. Simulate random walk of 20 steps in 1D. If the mean square displacement from origin is $<x^2>$,
### then show that $<x^2>$ ~ $t$ for large number of ensemble average.
### Here $t$ is the time steps of random walk.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
N = eval(input('Number of Ensemble: '))
steps = 20
X2 = []
steps_no = np.arange(20,110,10)

while steps<=100:
    X = []
    for i in range(N):
        x = 0
        for j in range(steps):
            p = random.random()
            if p >= 0.5:
                x += 1
            if p < 0.5:
                x -= 1
        X.append(x)
    #x_avg = sum(X)/N
    #print(x_avg)
    x2 = []
    for i in range(len(X)):
        d = X[i]**2
        x2.append(d)
    x2_avg = sum(x2)/N
    X2.append(x2_avg)
    print('When number of steps is '+str(steps)+' <x^2> = ',x2_avg)
    steps += 10
plt.plot(steps_no,X2)
plt.title('$ <x^2>$ ~ $t $')
plt.xlabel('time steps')
plt.ylabel('$<x^2>$')
plt.show()
```

### 7. Simulate random walk of 20 steps in 1D. Show the probability distribution for final position at 0,1... .

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
N = 50000
steps = 20
location = np.arange(-steps,steps+1)
fp = np.zeros(len(location))
X = []
for i in range(N):
	x = 0
	for j in range(steps):
		p = random.random()
		if p >= 0.5:
			x += 1
		if p < 0.5:
			x -= 1
	X.append(x)
	for k in range(len(location)):
		if location[k] == x:
			fp[k] += 1
print('Ensemble averaged over '+str(N)+' experiments')
x_avg = sum(X)/N
print('<x> :',x_avg)
X2 = []
for i in range(len(X)):
	d = X[i]**2
	X2.append(d)
x2_avg = sum(X2)/N
print('<x^2> :',x2_avg)
plt.bar(location,fp/N,color='r')
plt.xlabel('Position')
plt.ylabel('Probability')
plt.show()
```
