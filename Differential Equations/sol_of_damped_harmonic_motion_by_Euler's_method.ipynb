```python
# -*- coding: utf-8 -*-
"""
Created on Sun Dec 26 18:24:48 2021

@author: kd34w
"""
# Euler's Method for 2nd order ODE

# The ODE is : d2y/dx2 + lam*dy/dx + k*y = 0

import numpy as np
#Initial Conditions and increment in x
x0,y0,z0,h = 0,0,1,0.01
x = 20  #the x_coordinate in which we want the y value

# 2nd order DE into two 1st order coupled DE
def f(x,y,z): return z
def g(x,y,z):
    return -lam*z-k*y
lam,k = 0.5,2.0   #Parameters
y = y0
z = z0
X,Y,Z = [],[],[]
for i in np.arange(x0,x+h,h):
    X.append(i)
    Y.append(y)
    Z.append(z)
    print(i,y,z)
    y += h*f(i,y,z)
    z += h*g(i,y,z)

# for plotting
import matplotlib.pyplot as plt

plt.subplot(211)
plt.plot(X,Y)
plt.xlabel('Time')
plt.ylabel('Position')

plt.title('Damped Harmonic Motion')

plt.subplot(212)
plt.plot(X,Z,c='orange')
plt.xlabel('Time')
plt.ylabel('Velocity')

plt.show()
```
